```python
columns_to_drop = [
    'Customer ID', 'City', 'Zip Code', 'Latitude', 'Longitude',
    'Churn Category', 'Churn Reason'
]
df.drop(columns=columns_to_drop, inplace=True)
```

```python
numeric_coloumns = df.select_dtypes(include=['int64', 'float64']).columns
corr = df[numeric_coloumns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
def plot_histogram(df, column_name):
    plt.figure(figsize=(5, 3))

    data = pd.to_numeric(df[column_name], errors="coerce")

    sns.histplot(data, kde=True)
    plt.title(f"Distribution of {column_name}")

    col_mean = data.mean()
    col_median = data.median()

    plt.axvline(col_mean, color="red", linestyle="--", label=f"Mean: {col_mean:.2f}")
    plt.axvline(col_median, color="green", linestyle="-", label=f"Median: {col_median:.2f}")

    plt.legend()
    plt.show()

plot_histogram(df, "Tenure in Months")
plot_histogram(df, "Monthly Charge")
plot_histogram(df, "Total Charges")
```

```python
plt.pie(df['Customer Status'].value_counts(), labels=df['Customer Status'].unique(), autopct='%1.1f%%')
plt.title('Customer Status Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Customer Status', palette="Set2")
plt.title('Customer Status Distribution by Contract Type')
plt.xlabel("Contract Type")
plt.ylabel("Count")
plt.legend(title="Customer Status")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x ='Tenure in Months', hue='Customer Status', kde=True, element="step", stat="density", common_norm=False)
plt.title("Tenure Distribution by Customer Status")
plt.xlabel("Tenure (months)")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Internet Type', hue='Customer Status', palette="Set2")

plt.title("Internet Type Distribution by Customer Status")
plt.xlabel("Internet Type")
plt.ylabel("Count")
plt.legend(title="Customer Status")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Customer Status', y='Monthly Charge', palette="Set2")
plt.title("Monthly Charges by Customer Status")
plt.xlabel("Customer Status")
plt.ylabel("Monthly Charges")
plt.show()
```
